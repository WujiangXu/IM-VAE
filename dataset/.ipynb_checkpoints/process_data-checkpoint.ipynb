```
#overlap ratio 滑动 分train/test，train变，test不变
import os
import random
from typing import DefaultDict
from tqdm import tqdm
import numpy as np
import pandas as pd
from collections import defaultdict
import json
from random import sample

def select_overlap_user(df,save_train_name,overlap_ratio):
    data = df
    user_node = data['user_id'].tolist()
    seq_d1 = data['seq_d1'].tolist()
    seq_d2 = data['seq_d2'].tolist()
    domain_id = data['domain_id'].tolist()
    # also_view = data['also_view'].tolist()
    # also_view_d1 = data['also_view_d1'].tolist()
    # also_view_d2 = data['also_view_d2'].tolist()

    user_node_overlap,seq_d1_overlap, seq_d2_overlap, domain_id_overlap  = [], [], [], []
    user_node_nolap,seq_d1_nolap, seq_d2_nolap, domain_id_nolap  = [], [], [], []
    for i in range(len(user_node)):
        seq1_tmp = seq_d1[i]
        seq2_tmp = seq_d2[i]
        if len(seq1_tmp)!=0 and len(seq2_tmp)!=0:
            user_node_overlap.append(user_node[i])
            seq_d1_overlap.append(seq1_tmp)
            seq_d2_overlap.append(seq2_tmp)
            domain_id_overlap.append(domain_id[i])
            # also_view_overlap.append(also_view[i])
            # also_view_overlap_d1.append(also_view_d1[i])
            # also_view_overlap_d2.append(also_view_d2[i])
        else :
            user_node_nolap.append(user_node[i])
            seq_d1_nolap.append(seq1_tmp)
            seq_d2_nolap.append(seq2_tmp)
            domain_id_nolap.append(domain_id[i])
            # also_view_nolap.append(also_view[i])
            # also_view_nolap_d1.append(also_view_d1[i])
            # also_view_nolap_d2.append(also_view_d2[i])
    print(len(user_node_overlap),len(user_node_nolap)) # 3384 69945
    #nolap_num = int(len(user_node_overlap)/overlap_ratio-len(user_node_overlap)) # 3384 + 
    sample_overlap_num = int(len(user_node_overlap)*overlap_ratio)
    idx_lst = [i for i in range(len(user_node_overlap))]
    select_idx = sample(idx_lst, sample_overlap_num)
    print(sample_overlap_num)
    # print(select_idx)
    for i in select_idx:
        user_node_nolap.append(user_node_overlap[i])
        seq_d1_nolap.append(seq_d1_overlap[i])
        seq_d2_nolap.append(seq_d2_overlap[i])
        domain_id_nolap.append(domain_id_overlap[i])
        # also_view_nolap.append(also_view_overlap[i])
        # also_view_nolap_d1.append(also_view_overlap_d1[i])
        # also_view_nolap_d2.append(also_view_overlap_d2[i])

    # print(len(user_node_nolap))
    dataframe = pd.DataFrame({'user_id':user_node_nolap,'seq_d1':seq_d1_nolap,'seq_d2':seq_d2_nolap,'domain_id':domain_id_nolap})
    dataframe.to_csv(save_train_name,index=False,sep=',')

# data = df_all.sample(frac=1.0)
# train_len = int(data.shape[0] * 0.80)
# save_data_train = data.iloc[ : train_len]
# save_data_val = data.iloc[ train_len: ]
# train_name = "/ossfs/workspace/denoise_cdsr/dataset_14/phone_elec_train100.csv"
# val_name = "/ossfs/workspace/denoise_cdsr/dataset_14/phone_elec_test.csv"
# save_data_train.to_csv(train_name, index=False)
# save_data_val.to_csv(val_name, index=False)

save_data_train = pd.read_csv('cloth_sport_train.csv')
ratios = [0.25, 0.5, 0.75]
for overlap_ratio in ratios:
# #game video 90 refine
    save_train_name = "cloth_sport_train"+str(int(overlap_ratio*100))+".csv"
    select_overlap_user(save_data_train,save_train_name,overlap_ratio)
```

```
# generate cold-start users without also_view
from random import sample
import random
random.seed(2020)
def select_overlap_user_csu(df,save_train_name,overlap_ratio):
    data = df
    user_node = data['user_id'].tolist()
    seq_d1 = data['seq_d1'].tolist()
    seq_d2 = data['seq_d2'].tolist()
    domain_id = data['domain_id'].tolist()
    # also_view = data['also_view'].tolist()
    # also_view_d1 = data['also_view_d1'].tolist()
    # also_view_d2 = data['also_view_d2'].tolist()
    user_node_overlap,seq_d1_overlap, seq_d2_overlap, domain_id_overlap   = [], [], [], []
    user_node_nolap,seq_d1_nolap, seq_d2_nolap, domain_id_nolap  = [], [], [], []
    for i in range(len(user_node)):
        seq1_tmp = json.loads(seq_d1[i])
        seq2_tmp = json.loads(seq_d2[i])
        if len(seq1_tmp)!=0 and len(seq2_tmp)!=0:
            user_node_overlap.append(user_node[i])
            seq_d1_overlap.append(seq1_tmp)
            seq_d2_overlap.append(seq2_tmp)
            domain_id_overlap.append(domain_id[i])
        else :
            user_node_nolap.append(user_node[i])
            seq_d1_nolap.append(seq1_tmp)
            seq_d2_nolap.append(seq2_tmp)
            domain_id_nolap.append(domain_id[i])
    print(len(user_node_overlap),len(user_node_nolap)) # 3384 69945
    #nolap_num = int(len(user_node_overlap)/overlap_ratio-len(user_node_overlap)) # 3384 + 
    sample_overlap_num = int(len(user_node_overlap)*overlap_ratio)
    idx_lst = [i for i in range(len(user_node_overlap))]
    select_idx = sample(idx_lst, sample_overlap_num)
    print(sample_overlap_num)
    # print(select_idx)
    for i in range(len(user_node_overlap)):
        if i in select_idx: # cold-start users
            if domain_id_overlap[i]==0:
                user_node_nolap.append(user_node_overlap[i])
                seq_d1_nolap.append(seq_d1_overlap[i][-1:])
                seq_d2_nolap.append(seq_d2_overlap[i])
                domain_id_nolap.append(domain_id_overlap[i])
            else:
                user_node_nolap.append(user_node_overlap[i])
                seq_d1_nolap.append(seq_d1_overlap[i])
                seq_d2_nolap.append(seq_d2_overlap[i][-1:])
                domain_id_nolap.append(domain_id_overlap[i])  
        else:
            user_node_nolap.append(user_node_overlap[i])
            seq_d1_nolap.append(seq_d1_overlap[i])
            seq_d2_nolap.append(seq_d2_overlap[i])
            domain_id_nolap.append(domain_id_overlap[i])
    # print(len(user_node_nolap))
    dataframe = pd.DataFrame({'user_id':user_node_nolap,'seq_d1':seq_d1_nolap,'seq_d2':seq_d2_nolap,'domain_id':domain_id_nolap})
    dataframe.to_csv(save_train_name,index=False,sep=',')

save_data_val  = pd.read_csv("cloth_sport_train.csv")
ratio = [0.05, 0.35, 0.50]
for r in ratio:
    print('CSU ratio:', r)
    csu_name = "cloth_sport_train_csu_{}.csv".format(r)
    test_csu = select_overlap_user_csu(save_data_val,csu_name,r)
    print('\n')
```

```
# density experiment 
import pandas as pd 
def select_density_user(df_name,save_name,ratio):
    data = pd.read_csv(df_name,sep=',')
    user_node = data['user_id'].tolist()
    seq_d1 = data['seq_d1'].tolist()
    seq_d2 = data['seq_d2'].tolist()
    domain_id = data['domain_id'].tolist()
    # also_view = data['also_view'].tolist()
    # also_view_d1 = data['also_view_d1'].tolist()
    # also_view_d2 = data['also_view_d2'].tolist()

    user_node_overlap,seq_d1_overlap, seq_d2_overlap, domain_id_overlap  = [], [], [], []
    user_node_nolap,seq_d1_nolap, seq_d2_nolap, domain_id_nolap  = [], [], [], []
    for i in range(len(user_node)):
        seq1_tmp = json.loads(seq_d1[i])
        seq2_tmp = json.loads(seq_d2[i])
        if len(seq1_tmp)!=0:
            seq1_tmp = seq1_tmp[:max(2,int(len(seq1_tmp)*ratio))]
        if len(seq2_tmp)!=0:
            seq2_tmp = seq2_tmp[:max(2,int(len(seq2_tmp)*ratio))]
        user_node_nolap.append(user_node[i])
        seq_d1_nolap.append(seq1_tmp)
        seq_d2_nolap.append(seq2_tmp)
        domain_id_nolap.append(domain_id[i])
        # also_view_nolap.append(also_view[i])
        # also_view_nolap_d1.append(also_view_d1[i])
        # also_view_nolap_d2.append(also_view_d2[i])
    print(len(user_node_overlap),len(user_node_nolap)) # 3384 69945

    # print(len(user_node_nolap))
    dataframe = pd.DataFrame({'user_id':user_node_nolap,'seq_d1':seq_d1_nolap,'seq_d2':seq_d2_nolap,'domain_id':domain_id_nolap})
    dataframe.to_csv(save_name,index=False,sep=',')

density_ratio = [0.25, 0.5, 0.75]
df_name = "cloth_sport_test_csu.csv"
for dr in density_ratio:
    save_name = "cloth_sport_test_csuDS{}.csv".format(int(dr*100))
    select_density_user(df_name,save_name,dr)
```
